```python
from sentence_transformers import SentenceTransformer
import numpy as np

def cosine_similarity(vec1, vec2):
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

model = SentenceTransformer('paraphrase-distilroberta-base-v1')

tags = ['low quality', 'sustained strings melody', 'soft female vocal', 'mellow piano melody', 'sad', 'soulful', 'ballad']

your_caption = "The low quality recording features a ballad song that contains sustained strings, mellow piano melody and soft female vocal singing over it. It sounds sad and soulful, like something you would hear at Sunday services."
musiccaps_caption = "The song features a mellow piano melody and sustained strings that create a melancholic and soulful atmosphere, perfectly complementing the sad and emotional ballad style. A soft female vocal further enhances the poignant mood, delivering heartfelt and sincere expressions of sorrow. Despite its 'low quality' tag, the unique combination of these musical elements results in a deeply moving and evocative listening experience"

# Concatenate tags into a single sentence
tags_sentence = ' '.join(tags)

# Compute embeddings
tags_embedding = model.encode(tags_sentence)
your_caption_embedding = model.encode(your_caption)
musiccaps_caption_embedding = model.encode(musiccaps_caption)

# Calculate cosine similarities
your_caption_similarity = cosine_similarity(tags_embedding, your_caption_embedding)
musiccaps_caption_similarity = cosine_similarity(tags_embedding, musiccaps_caption_embedding)

print(f"Similarity between the tags and your caption: {your_caption_similarity:.2f}")
print(f"Similarity between the tags and musicCaps caption: {musiccaps_caption_similarity:.2f}")

if your_caption_similarity > musiccaps_caption_similarity:
    print("Your caption is more semantically aligned with the tags.")
elif your_caption_similarity < musiccaps_caption_similarity:
    print("The musicCaps caption is more semantically aligned with the tags.")
else:
    print("Both captions are equally semantically aligned with the tags.")
```
