```python
from google.colab import drive
drive.mount('/content/gdrive')

import numpy as np 
import os
import json 
import pandas as pd

ROOT_DATA = "/content/gdrive/MyDrive/IDL_Project/data"
AUDIOSET="/content/gdrive/MyDrive/IDL_Project/data/audioset"
audioset_paths = {
    "eval": os.path.join(AUDIOSET,"csvs","eval_segments.csv"),
    "eval_tagged": os.path.join(AUDIOSET,"tags","eval_tagged.csv"),
    "eval_caps": os.path.join(AUDIOSET,"captions","eval_caps.csv"),
    "music": os.path.join(AUDIOSET,"sounds","Music")
}
```

```python
caps_data = {}
with open(audioset_paths["eval_caps"], "r") as file:
  for line in file.readlines():
    info = line.split(",")
    caps_data[info[0]] = {
        "time": info[1:3],
        "caps": " ".join(info[3:])}
caps_data.keys()
```

```python
tags = pd.read_csv(audioset_paths["eval_tagged"], header=None)
tags.head()
```

```python
from IPython.display import Audio, display
import random
def get_random_file():
  choice = random.choice(list(caps_data.items()))
  return (choice[0], choice[1]["time"][0])

def display_audio(sound_id=None, autoplay=True):
  if sound_id is None:
    sound_id = get_random_file()
  
  sound_file = os.path.join(audioset_paths["music"], f"{sound_id[0]}_{sound_id[1]}.wav")
  print(sound_id[0])
  print(caps_data[sound_id[0]]["time"])
  print(tags[tags[0]==sound_id[0]][3].item())
  print(caps_data[sound_id[0]]["caps"].replace('" "', "\n"))
  
  display(Audio(sound_file, autoplay=autoplay))
```

```python
for i in range(5): 
  display_audio(sound_id=None, autoplay= False)
```
